```python
quick_bites_df["reviews_list"] = quick_bites_df["reviews_list"].apply (lambda x:x.lower())
```

```python
from nltk.corpus import RegexpTokenizer
```

```python
tokenizer = RegexpTokenizer("[a-zA-Z]+")
```

```python
tokenizer.tokenize(quick_bites_df["reviews_list"][3])
```

```python
sample = data[0:100000]
```

```python
reviews_tokens = sample["reviews_list"].apply(tokenizer.tokenize)
```

```python
reviews_tokens
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stop = stopwords.words("english")
```

```python
print(stop)
```

```python
stop.extend(['a', 'n', 'nan', 'RATED', 'Rated', 'x', 'f', 'I', 'xa', 'b', 'c' ])
```

```python
print(stop)
```

```python
rev3 = reviews_tokens[3]
```

```python
print ([token for token in rev3 if token not in stop])
```

```python
token_review_clean = reviews_tokens.apply (lambda each_review : [token for token in each_review if token not in stop])
```

```python
token_review_clean
```

```python
total_review_clean_2d = list(token_review_clean)  #converted to 2d list)
```

```python
total_review_clean_1d = []
for review in total_review_clean_2d:
 for word in review:
    total_review_clean_1d.append(word)    
```

```python
total_review_clean_1d 
```

```python
from nltk import FreqDist #  unigram analysis
```

```python
fd = FreqDist()
```

```python
for word in total_review_clean_1d:
    fd[word] = fd[word] + 1
```

```python
fd.most_common(20)
```

```python
fd.plot(20)
```

Bigram & Trigram

```python
from nltk import bigrams, trigrams
```

```python
bi_grams = bigrams(total_review_clean_1d)
```

```python
bi_grams
```

```python
fd_bigrams = FreqDist()
for bigrams in bi_grams:
    fd_bigrams[bigrams] = fd_bigrams[bigrams] + 1
```

```python
fd_bigrams.most_common(50)
```

```python
fd_bigrams.plot(50)
```

```python
tri_grams = trigrams(total_review_clean_1d)
```

```python
fd_trigrams = FreqDist()
for trigrams in tri_grams:
    fd_trigrams[trigrams] = fd_trigrams[trigrams] + 1
```

```python
fd_trigrams.most_common(50)
```

```python
fd_trigrams.plot(50)
```
